```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("intersection.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
crash_data = data[['CRASH TIME','coordinates']].iloc[:2000,:]
crash_data
```

```python
crash_data.iloc[0,1]
```

```python
lat = []
lon = []



for crash in crash_data['coordinates']:
    
    lat.append(crash[1:-1].split(',')[0])
    lon.append(crash[1:-1].split(',')[1])
    
    
crash_data['latitude'] = lat
crash_data['longitude'] = lon
    
```

```python
crash_data.drop(columns=['coordinates'], inplace=True)
```

```python
crash_data.to_csv('intersection_data_for_fire_base_all.csv')
```
